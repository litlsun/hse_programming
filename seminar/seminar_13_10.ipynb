```python
print("=== ФИНАНСОВЫЙ ОТЧЕТ ===")

company = "ООО 'Рога и копыта'"
quarter = 4
year = 2023
revenue = 15000000.5678
expenses = 9000000.1234
profit = revenue - expenses
profit_margin = (profit / revenue) * 100

report = f"""
{company.upper()}
ОТЧЕТ ЗА {quarter}-Й КВАРТАЛ {year} ГОДА
--------------------------------------------------
Выручка:            {revenue:15,.2f} ₽
Расходы:            {expenses:15,.2f} ₽
Прибыль:            {profit:15,.2f} ₽
Рентабельность:     {profit_margin:15.1f} %
"""

print(report)
```

```python
print("=== ПРОГРЕСС ВЫПОЛНЕНИЯ ЗАДАЧ ===")

total_tasks = 150
completed_tasks = 87
progress = (completed_tasks / total_tasks) * 100

bar_length = 30
filled_length = int(bar_length * completed_tasks // total_tasks)
bar = '█' * filled_length + '░' * (bar_length - filled_length)

report = f"""
ВЫПОЛНЕНИЕ ЗАДАЧ:
[{bar}] {progress:.1f}%
Завершено: {completed_tasks}/{total_tasks} задач
"""

print(report)
```

```python
text = "   Python - отличный язык для начинающих! Python мощный.   "

# 1. Уберите лишние пробелы
clean_text = text.strip()

# 2. Подсчитайте общее количество символов
total_chars = len(clean_text)

# 3. Подсчитайте сколько раз встречается слово "Python"
python_count = clean_text.count("Python")

# 4. Проверьте, начинается ли текст с "Python"
starts_with_python = clean_text.startswith("Python")

# 5. Переведите текст в нижний регистр
lower_text = clean_text.lower()

print(f"Очищенный текст: '{clean_text}'")
print(f"Всего символов: {total_chars}")
print(f"Слов 'Python' найдено: {python_count}")
print(f"Начинается с 'Python': {starts_with_python}")
print(f"В нижнем регистре: {lower_text}")
```

```python
user_data = "ivanov:Иван:Петров:25:Москва"

# 1. Разбиваем строку по разделителю ":"
data_parts = user_data.split(":")

# 2. Извлекаем отдельные данные
username = data_parts[0]
first_name = data_parts[1]
last_name = data_parts[2]
age = data_parts[3]
city = data_parts[4]

print(f"Логин: {username}")
print(f"Имя: {first_name}")
print(f"Фамилия: {last_name}")
print(f"Возраст: {age}")
print(f"Город: {city}")
```
