#Наборы данных

Коллекции (наборы) данных - это структуры в Python, которые позволяют хранить несколько элементов в одной переменной

- List (Список) - упорядоченный, изменяемый
- Tuple (Кортеж) - упорядоченный, неизменяемый
- Set (Множество) - неупорядоченный, хранит уникальные элементы

```python
text = "программирование"

print(set(text))
```

```python
text = 'university'

print(set(text))
print(list(text))
print(tuple(text))
```

Задание 2.

Дано:

set1 = {1, 2, 3, 4, 5}

set2 = {4, 5, 6, 7, 8}

Найдите:
1. Общие элементы (пересечение)
2. Все уникальные элементы (объединение)
3. Элементы, которые есть в set1, но нет в set2

```python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

both_sets = set1 | set2
unique_elements = set1 & set2
unique_set1 = set1 - set2
print(f'Общие элементы: {both_sets}\nВсе уникальные элементы: {unique_elements}\nЭлементы, которые есть в set1, но нет в set2: {unique_set1}')
```

Задание 3

```python
mixed_data = [
    1,
    "hello",
    3.14,
    [1, 2],
    42,
    "world",
    0,
    (5, 6),
    {"name": "John"},
    True,
    {7, 8, 9}
]

int_list = []
str_list = []
list_list = []
tuple_list = []
float_list = []
set_list = []
bool_list = []
other_list = []

for item in mixed_data:
   if isinstance(item, bool):
       (bool_list.append(item))
   elif isinstance(item, int):
       (int_list.append(item))
   elif isinstance(item, str):
       (str_list.append(item))
   elif isinstance(item, list):
       (list_list.append(item))
   elif isinstance(item, tuple):
       (tuple_list.append(item))
   elif isinstance(item, float):
       (float_list.append(item))
   elif isinstance(item, set):
       (set_list.append(item))
   else:
       (other_list.append(item))

print(f"Целые числа: {int_list}\nСтроки: {str_list}\nСписки: {list_list}\nКортежи: {tuple_list}\nДробные числа: {float_list}\nМножества: {set_list}\nЛогические значения: {bool_list}\nДругое: {other_list}")
```
