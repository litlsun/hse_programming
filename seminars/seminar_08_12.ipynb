```python
student1 = {
    "id": 101,
    "name": "Анна Иванова",
    "age": 20,
    "faculty": "Информатика",
    "grades": {"Программирование": 95, "Математика": 88, "Базы данных": 92}
}

student2 = dict(
    id=102,
    name="Петр Сидоров",
    age=22,
    faculty="Физика",
    grades={"Механика": 78, "Оптика": 85, "Квантовая физика": 90}
)

student3 = dict([
    ("id", 103),
    ("name", "Мария Петрова"),
    ("age", 21),
    ("faculty", "Экономика"),
    ("grades", {"Макроэкономика": 91, "Статистика": 87, "Финансы": 89})
])

print(student1)
print(student2)
print(student3)
```
