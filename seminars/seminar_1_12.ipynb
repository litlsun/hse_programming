```python
# Дан список слов. Найдите все слова, содержащие букву 'а', и замените их на "FOUND"

words = ["apple", "banana", "cherry", "date", "fig", "grape"]
for i in range(len(words)):
    if 'a' in words[i]:
        words[i] = "FOUND"

print(words)
```

```python
#У вас есть данные о продажах: товары, цены и количество проданных единиц

products = ["Ноутбук", "Мышь", "Клавиатура", "Монитор"]
prices = [50000, 2500, 4000, 30000]
quantities = [8, 45, 25, 12]

#Рассчитайте общую выручку по каждому товару

total_revenue_by_product = [0,0,0,0]
for i in range(len(products)):
    total_revenue_by_product[i] = prices[i] * quantities[i]

print(total_revenue_by_product)

# и найдите товар с максимальной выручкой

for i in range(len(products)):
    total_revenue_by_product[i] = prices[i] * quantities[i]

max_revenue = max(total_revenue_by_product)
index_max = total_revenue_by_product.index(max_revenue)
product_max = products[index_max]

print("Товар с максимальной выручкой:", product_max)
print("Выручка:", max_revenue)
```

```python
# В интернет-магазине нужно применить скидки к товарам, обновить цены и пересчитать остатки (просто вывести значение stock по индексу)

products = ["Телефон", "Планшет", "Ноутбук", "Наушники"]
prices = [30000, 45000, 80000, 15000]
stock = [15, 8, 5, 20]
discounts = [10, 15, 20, 5]  # Скидки в процентах

for i in range(len(products)):
    new_price = prices[i] * (100 - discounts[i]) / 100
    print(f"{products[i]}: старая цена {prices[i]}, новая цена {new_price}, остаток {stock[i]}")
```
