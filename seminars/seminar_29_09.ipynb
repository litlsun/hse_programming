```python
# Это пример первой программы на Python.
# Она выводит на экран приветственное сообщение "Hello, World!".
# В комментариях объясняется назначение кода.

def main():
    """
    Главная функция программы.

    Функция выводит строку "Hello, World!" в консоль.
    """
    print("Hello, World!")  # вывод текста на экран
```

```python
# Точка входа в программу.
# Код в этом блоке выполнится только если файл запущен напрямую.
if __name__ == "__main__":
    main()
```

```python
#Оператор += совмещает присваивание со сложением / конкатенацией для скорости написания кода

text = 'Hello, World!'
text += 'Wow!' # конкатенация строки в text со строкой “Wow!”
print(text)
```

4. Записать в переменную name имя Анна. Записать в переменную surname фамилию Смирнова. Вывести приветствие с помощью +=

```python
# Ожидаемый вывод: Привет, Анна Смирнова!

name = "Анна"
surname = "Смирнова"

greeting = "Привет, "
greeting += name + " " + surname + "!"

print(greeting)
```

   Запишите в переменные name, surname, age данные пользователя
   Анна Смирнова, которой 25 лет. Выведите результат, используя +=
   
   Hint: вызов print() с новой строки, создает новую строку в выводе
   print("Строка один")
   print("Строка два")

   Результат в консоли:
   Строка один
   Строка два

   Еще для создания новой строки можно использовать "\n" (ньюлайн).
   "\n" работает как обычная строка, ее можно конкатенировать,
   передавать через запятую, просто "принтить"

```python
# Ожидаемый вывод:
# User: Анна Смирнова
# Age: 25

name = "Анна"
surname = "Смирнова"
age = 25

first_line = "User: "
first_line += name + " " + surname

second_line = "Age: "
second_line += str(age)

print(first_line)
print(second_line)
```
