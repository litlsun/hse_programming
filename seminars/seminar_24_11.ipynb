# Задача 1 (Базовый while и if)

Вы пишете программу для проверки сложности пароля. Пароль должен быть не короче 8 символов.

Напишите программу, которая постоянно запрашивает у пользователя пароль с помощью input(). Если введённый пароль короче 8 символов, программа выводит сообщение "Пароль слишком короткий. Попробуйте снова." и продолжает запрашивать пароль. Как только пользователь введёт пароль длиной 8 или более символов, программа выводит "Пароль принят!" и завершает работу.


```python
while True:
    password = input("Введите пароль: ")

    if len(password) >= 8:
        print("Пароль принят!")
        break
    else:
        print("Пароль слишком короткий. Попробуйте снова.")
```

# Задача 2 (while, if и tuple)

Вы создаёте простой интерфейс для выбора действия из фиксированного меню.

Используйте tuple для хранения доступных команд ('start', 'stop', 'pause', 'quit'). Программа должна в цикле запрашивать у пользователя команду. Если команда есть в кортеже, программа выводит сообщение "Выполняю команду: [команда]". Если команда 'quit', программа выводит "Выход из программы." и завершает работу. Если команды нет в списке, выводится "Неизвестная команда."


```python
commands = ('start', 'stop', 'pause', 'quit')

while True:
    user_command = input("Введите команду: ").lower()
    if user_command in commands:
        print(f"Выполняю команду: {user_command}")
        if user_command == 'quit':
            print("Выход из программы.")
            break
    else:
        print("Неизвестная команда.")
```

# Задача 3 (while, if и list)

Вы помогаете другу вести список покупок.

Создайте пустой список shopping_list. Программа должна предлагать пользователю на выбор: 1) Добавить товар, 2) Удалить товар, 3) Показать список, 4) Выйти. В зависимости от вывода пользователя программа выполняет соответствующее действие с списком shopping_list с помощью методов .append() и .remove(). Цикл продолжается, пока пользователь не выберет вариант "Выйти".


```python
shopping_list = []

while True:
    print("1) Добавить товар")
    print("2) Удалить товар")
    print("3) Показать список")
    print("4) Выйти")

    choice = input("Выберите действие: ")

    if choice == '1':
        item = input("Введите товар: ")
        shopping_list.append(item)

    elif choice == '2':
        item = input("Введите товар для удаления: ")
        if item in shopping_list:
            shopping_list.remove(item)

    elif choice == '3':
        print("Список покупок:", shopping_list)

    elif choice == '4':
        break
```

# Задача 4 (while, if и set)

Вы пишете программу для сбора уникальных email-адресов для рассылки.

Создайте пустое множество email_set. Программа должна запрашивать у пользователя email-адреса по одному. При вводе адреса он добавляется в множество (множество автоматически игнорирует дубликаты). Если пользователь вводит пустую строку, программа прекращает сбор адресов и выводит все уникальные email-адреса в отсортированном порядке.


```python
email_set = set()

while True:
    email = input("Введите email-адрес (или нажмите Enter для завершения): ")

    if email == "":
        break

    email_set.add(email)

print("Уникальные email-адреса:")
for email in sorted(email_set):
    print(email)
```

# Задача 5 (while, if, list comprehension)

Вы анализируете данные с датчика, который периодически отправляет целые числа. Вам нужно собирать только чётные положительные числа, пока не будет получен 0 (сигнал остановки).

Напишите программу, которая в цикле while запрашивает целое число. Используйте list comprehension внутри цикла, чтобы сразу фильтровать ввод: добавлять в итоговый список только те числа, которые являются чётными и положительными. Цикл должен завершиться при вводе числа 0. В конце программа выводит результирующий список и его длину.


```python
numbers = []

while True:
    num = int(input("Введите целое число (0 для остановки): "))

    if num == 0:
        break

    # Используем list comprehension для фильтрации
    filtered = [x for x in [num] if x > 0 and x % 2 == 0]
    numbers.extend(filtered)

print("Результат:", numbers)
print("Длина списка:", len(numbers))
```
