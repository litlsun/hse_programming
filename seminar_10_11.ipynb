Условие:
Интернет-магазин предоставляет скидку 10% при выполнении одного из условий:

    Сумма покупки ≥ 5000 рублей ИЛИ

    Пользователь имеет действующий промокод

Напишите программу, которая рассчитывает итоговую цену с учетом скидки.

Пример входных данных:

    Сумма покупки: 6000 рублей

    Промокод: False

```python
purchase_amount = 6000
has_promo_code = False

if purchase_amount >= 5000 or has_promo_code:
    discount = purchase_amount * 0.1
    final_price = purchase_amount - discount
    print(f"Итоговая цена: {final_price} рублей")
    print(f"Скидка: {discount} рублей")
else:
    print(f"Итоговая цена: {purchase_amount} рублей")
    print("Скидка не применяется")
```

Задача 5. Фильтрация комментариев на сайте

Условие:
Модерационная система одобряет комментарий если:

    Длина комментария ≤ 500 символов И

    В комментарии отсутствуют запрещенные слова

Напишите программу, которая проверяет, можно ли опубликовать комментарий.

Пример входных данных:

    Комментарий: "Отличный пост! Спасибо за информацию."

    Запрещенные слова: ["спам", "реклама", "вирус"]

```python
comment = "Отличный пост! Спасибо за информацию. Спам!"
banned_words = ["спам", "реклама", "вирус"]

has_valid_length = len(comment) <= 500
has_no_banned_words = not any(banned_word in comment.lower() for banned_word in banned_words)

if has_valid_length and has_no_banned_words:
    print("Комментарий: ОДОБРЕН для публикации")
else:
    print("Комментарий: НЕ ОДОБРЕН для публикации")
    if not has_valid_length:
        print(f"- Превышена длина ({len(comment)} > 500 символов)")
    if not has_no_banned_words:
        found_banned = [word for word in banned_words if word in comment.lower()]
        print(f"- Содержит запрещенные слова: {', '.join(found_banned)}")
```

Задача 1. Простой анализ текста

Условие:
Виртуальный ассистент анализирует сообщения пользователя,
чтобы определить приоритет и тип задачи.

Система должна:

    Вывести на экран сообщение "high priority", есть текст
    содержит слово "срочно" ИЛИ длиннее 50 символов

    Вывести "question" если содержит знак вопроса

    В остальных случаях -> вывести длину сообщения

Пример входных данных:

и предлагает альтернативы.

Логика перенаправления:

    Если уровень пробок выше 7 И есть альтернативный маршрут ->
    искать объезд

    Если до цели меньше 10 км -> показывать оставшееся время

    Если пробки выше 5 И нет альтернативы -> предупредить о пробках

    В остальных случаях продолжать по текущему маршруту

Напишите программу, которая принимает решение о смене маршрута.

Пример входных данных:

    Уровень пробок: 8/10
    Оставшееся расстояние: 25 км
    Оставшееся время: 45 минут
    Есть альтернативный маршрут: True


```python
traffic_level = 8
distance_left = 25
time_left = 45
has_alternative = True

if traffic_level > 7 and has_alternative:
    print("искать объезд")
elif distance_left < 10:
    print(f"показывать оставшееся время - {time_left} минут")
elif traffic_level > 5 and not has_alternative:
    print("предупредить о пробках")
else:
    print("продолжать по текущему маршруту")
```

Задача 3. Тернарный оператор с логическим И (and)

Условие:
Напишите программу, которая определяет скидку используя тернарный оператор с условием AND.

Логика:

    Если сумма покупки > 5000 И есть промокод -> скидка 15%
    Иначе -> скидка 5%

Пример входных данных:

    Сумма покупки: 6000
    Есть промокод: True

```python
purchase_amount = 6000
has_promo_code = True

discount = 0.15 if purchase_amount > 5000 and has_promo_code else 0.05

final_price = purchase_amount * (1 - discount)

print(f"Скидка: {discount * 100}%")
print(f"Итоговая цена: {final_price:.0f} рублей")
```

Задача 4. Тернарный оператор с регулярными выражениями

Условие:
Напишите программу, которая проверяет email на валидность используя тернарный оператор и regex.

Логика:

    Если email содержит '@' и '.' -> "Валидный email"
    Иначе -> "Невалидный email"

Пример входных данных:

    Email: "user@example.com"

```python
import re
email = "user@example.com"
result = "Валидный email" if re.match(r'^[^\s@]+@[^\s@]+\.[^\s@]+$', email) else "Невалидный email"
print(result)
```

Задача 5. Сложная проверка с регулярными выражениями

Условие:
Напишите программу, которая анализирует сложность пароля используя тернарный оператор и regex.

Логика:

    Если пароль содержит цифры И
    буквы в верхнем и нижнем регистре И длиной >= 8 -> "Надежный пароль"
    Иначе если пароль содержит цифры ИЛИ буквы в разных регистрах ->
    "Средний пароль"  
    Иначе -> "Слабый пароль"

Пример входных данных:

    Пароль: "Secure123"

```python
password = "Secure123"

has_digit = bool(re.search(r'\d', password))
has_upper = bool(re.search(r'[A-Z]', password))
has_lower = bool(re.search(r'[a-z]', password))
has_min_length = len(password) >= 8

strength = ("Надежный пароль" if has_digit and has_upper and has_lower and has_min_length else
            "Средний пароль" if has_digit or (has_upper and has_lower) else
            "Слабый пароль")

print(strength)
```
